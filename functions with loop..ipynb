```python
def validate(amount):
    if amount < 50:
        return False
    if amount not in [199, 299, 399, 599]:
        return False
    return True

while True:
    amt = int(input("Enter recharge amount: "))

    if validate(amt):
        print("Recharge successful")
        break
    else:
        print("Invalid amount, try again")
```

```python
def check_stock(inventory):
    for item in inventory:
        if inventory[item] < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")
products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 30
}

check_stock(products)
```

```python
def result(marks):
    total = 0
    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [60, 55, 40, 50]
print(result(marks))

```

```python
def calculate_fare(km, peak):
    fare = 50 + (km * 12)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare


while True:
    distance = int(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    print("Total Fare: ₹", calculate_fare(distance, peak_hour))

    retry = input("Retry? (yes/no): ")
    if retry != "yes":
        break
```

```python
def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "A", "P", "P", "P", "A"]
print(check_eligibility(attendance_list))
```

```python
def check_password(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
print(check_password(pwd))
```
